```python
from scipy.stats import norm
import numpy as np
import random
import matplotlib.pyplot as plt
from math import *
```

# Random de prueba

```python
mu_rand, sigma, n = random.randint(100, 101), 4, 16 #Datos
sample = np.random.normal(mu_rand, sigma, n)       #Muestra
mean_s = np.mean(sample)                           #Prom. muestra

#-------

mu, alph = 100, 0.05

a = (mean_s - mu) / (sigma / n**0.5)
p_value = (1 - norm.cdf(a))
print(p_value)

cr_region = norm.ppf(1-alph)
print(cr_region)

prop = a >= cr_region
print(prop)
```

# Ploteo

```python
l = 3.0
def f(x):
    return (exp( -x**2/2 ))/sqrt(2*pi)
x = np.arange(-l, l, 0.01)
n = np.shape(x)[0]
y = np.zeros(n)
for i in range(n): y[i] = f(x[i])
z = np.zeros(n)
for i in range(n):
    if x[i] >= cr_region:
        z[i] = y[i]

fig, ax = plt.subplots(figsize= (10,5))

ax.plot(x, y)
#ax.axis('equal')
#ax.fill_between(x, 0, y)
ax.fill_between(x, 0, z)


ax.annotate(
    'RegiÃ³n critica = ' + str(round(cr_region, 3)), xy=(cr_region, 0.05), xytext=(0, 0.1),
    arrowprops=dict(facecolor='black', shrink=0.05)
)

fig.savefig('graficos\grafico_1.png')

plt.show()
```
